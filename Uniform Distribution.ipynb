# Uniform Distribution

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
from scipy import stats
```

```python
un = stats.uniform.rvs(loc=1, scale=10, size=1000)
un
```

```python
sns.histplot(un)
```

```python
sns.histplot(un,bins=50)
```

```python
sns.histplot(un,bins=50,stat='density')
```

```python
sns.distplot(un)
```

```python
sns.histplot(un,kde= True,stat="density")
```

```python
  sns.kdeplot(un)
```

```python
stats.uniform.pdf(-10, loc=1, scale=100)
```

```python
stats.uniform.cdf(20, loc=1, scale=100)
```

```python
x=np.arange(-40,140,1)
x
```

```python
cdf_un = stats.uniform.cdf(x,loc=1 , scale=100)
cdf_un
plt.plot(cdf_un)
```
